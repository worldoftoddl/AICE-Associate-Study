## 데이터 선택하기

### 1. 컬럼명으로 데이터 선택

```python
# CSV 파일 읽기
file_path = 'datasets/winequality-white.csv'
wine_data = pd.read_csv(file_path, sep=",")     ## read_csv이기 때문에 sep = ','는 안써도 됨

# 'quality' 열 선택
quality_data = wine_data[['quality']]

# 데이터 확인
quality_data.head()
```

```python
# 'quality'와 'pH' 열 선택
selected_columns = wine_data[['quality', 'pH']]
selected_columns.head()
```

### 2. 범위로 컬럼 선택

```python
# 데이터프레임의 상위 10개 행 선택
top_10_rows = wine_data[:10]
top_10_rows
```

### 3. 특정 열과 행 선택

```python
# iloc (integer location) - 행과 열을 정수로 표현
value = wine_data.iloc[0, 1]
print(value)

# loc (label location) - 행의 이름(label)과 열 번호로 표현
value1 = wine_data.loc[0, 'volatile acidity']
print(value1)
```

```python
### 특정 열과 범위 행 선택
my_rows = wine_data[['quality','pH']][0:20]
my_rows
```

```python
## 주의할 점
print(wine_data[['quality']])  # Series가 아니라 DataFrame 반환
print(wine_data['quality'])    # 그냥 series 반환
```

### 4. 특정 조건을 만족시키는 데이터 선택

```python
# quality가 7이상인 데이터 선택
high_quality = wine_data[wine_data['quality'] >= 7]     # w[w['q'] >= 7]  ==> w[ T / F ] ==> T인 열들을 모은 dataframe 
high_quality.head()
```

```python
# quality가 7 이상이고 pH가 3 이상인 데이터
fine_quality = wine_data[(wine_data['quality']>=7) & (wine_data['pH'] >= 3)]        ## DataFrame에서는 and 쓰면 안된다~
fine_quality.head()

fine_quality.to_csv('test.csv', index = False)
```
